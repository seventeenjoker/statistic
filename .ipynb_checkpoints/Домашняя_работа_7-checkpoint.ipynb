```python
import numpy as np
import matplotlib.pyplot as plt

def mse(x, y, b1, n):
    return np.sum((b1*x-y)**2)/n
```

### 1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Используя математические операции, посчитать коэффициенты линейной регрессии, приняв за X заработную плату (то есть, zp - признак), а за y - значения скорингового балла (то есть, ks - целевая переменная). Произвести расчет как с использованием intercept, так и без.

```python
# количество наблюдений

x = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

n = len(x)
n
```

```python
b = (np.mean(x * y) - np.mean(x) * np.mean(y)) / (np.mean(x**2) - np.mean(x) ** 2)
b
```

```python
a = np.mean(y) - b * np.mean(x)
a
```

Итак, уравнение регрессии имеет вид (коэффициенты округлены до сотых):

$$y = 444.2 + 2.6 \cdot x$$

Найдем коэффициенты линейной регрессии применив Матричный метод

```python
X = x.reshape(len(x), 1)
X
```

```python
y = y.reshape(len(y), 1)
y
```

```python
X_1 = np.hstack([np.ones((10,1)), X])
X_1
```

```python
B_1 = np.dot(np.linalg.inv(np.dot(X_1.T, X_1)),X_1.T@y)
print(f'a = {B_1[0][0]}; b = {B_1[1][0]}')
```

Произведем расчет коэффициента линейной регрессии Матричным методом без intercept

```python
B = np.dot(np.linalg.inv(np.dot(X.T, X)),X.T@y)
print(f'b = {B[0][0]}')
```

```python
plt.rcParams['figure.figsize'] = [20, 8]
plt.scatter(X, y)
plt.plot(X, a+b*X, c='r', label='Линейная регрессия с intercept')
plt.plot(X, B[0][0]*X, c='b', label='Линейная регрессия без intercept')
plt.xlabel('Заработная плата', fontsize=24)
plt.ylabel('Кредитный скоринг', fontsize=24)
plt.legend(fontsize=24)
plt.show()
```

### 2. Посчитать коэффициент линейной регрессии при заработной плате (zp), используя градиентный спуск (без intercept).

```python
b1 = 0.1
n = len(x)
alpha = 1e-6
alpha
```

```python
for i in range(10):
    b1 -= alpha*(2/n)*np.sum((b1*x-y)*x)
    print(f"B1 = {b1}")
```

```python
for i in range(100):
    b1 -= alpha*(2/n)*np.sum((b1*x-y)*x)
    if i%10 == 0:
        print(f"Iteration: {i}, B1 = {b1}, mse = {mse(x, y, b1, n)}")
```

```python
for i in range(3000):
    b1 -= alpha*(2/n)*np.sum((b1*x-y)*x)
    if i%500 == 0:
        print(f"Iteration: {i}, B1 = {b1}, mse = {mse(x, y, b1, n)}")
```

```python
mse(x, y, 5.889820420132673, n)
```

```python
print(f'Iteration: 2500, B1 = {5.889820420132673}, mse = {56516.85841571943}')
```

### 3. Произвести вычисления как в пункте 2, но с вычислением intercept. Учесть, что изменение коэффициентов должно производиться на каждом шаге одновременно (то есть изменение одного коэффициента не должно влиять на изменение другого во время одной итерации).

```python
def mse_2(A2, B2, x=x, y=y):
  return (((A2+B1*x)-y)**2).mean()
```

```python
plt.rcParams['figure.figsize'] = [20, 10]
plt.scatter(x, y)
plt.xlabel('Заработная плата', fontsize=24)
plt.ylabel('Кредитный скоринг', fontsize=24)

A2 = 1000
B1 = 0.1
B2 = 0.1
mse2 = 0
plt.plot(x, A2+B2*x, c='g')
for i in range(1000000):
  A2 -= 100*alpha*2*((A2+B2*x)-y).mean()
  B2 -= alpha*2*(((A2+B2*x)-y)*x).mean()
  mse__2 = mse_2(A2, B2)
  if abs(mse__2-mse2) < alpha:
    print(f'Iteration: {i}, A2 = {A2}, B2 = {B2}, mse = {mse}')
    break
  mse2 = mse__2
  if i%20000==0:
    plt.plot(x, A2+B2*x, c='r')
plt.plot(x, A2+B2*x, c='b')
print(f'Сравним с ранее вычисленными: a = {B_1[0][0]}; b = {B_1[1][0]}')
```
