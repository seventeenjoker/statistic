```python
import pandas as pd
import numpy as np
```

```python
def mid(a, b):
    return (a + b) / 2
def dispersia(a, b):
    return (b - a)**2/12
```

## 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800] Найдите ее среднее значение и дисперсию.

```python
# среднее
print(mid(200, 800))
```

```python
# дисперсия
print(dispersia(200, 800))
```

## **Ответ:** среднее - 500, дисперсия 30000.

## 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

По следующей формуле можно рассчитать дисперсию $D(X)$:


$$D(X) = \frac{(b - a)^2}{12}$$

```python
Dx = 0.2
a = 0.5
```

```python
# Тогда выведем из этой формулы b
b = np.sqrt(Dx*12) + a
b
```

```python
b = 2.0492
# тогда среднее
print(mid(a, b))
```

## **Ответ:** можно, правая граница - b = 2.0492, среднее - 1.2746.

## 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения

$$f(x) = \Bigl(\frac{1} {4 * \sqrt{2\pi}}\Bigr) * \exp^{-\frac{(x+2)^2}{ 32}}$$

Найдите:

а). M(X)

б). D(X)

в). std(X) (среднее квадратичное отклонение)

Общий вид: $$f(x) = \frac{1}{\sigma \sqrt{2 \pi}} e ^ {- \frac{(x - a)^2}{2 \sigma^2}},$$
где  $$a=M(X),σ2=D(X)$$
#### таким образом

### $$D(X) = 4^2$$
### $$std(X) = 4$$
